```python
import pandas as pd
```

```python
dengue = pd.read_csv("dengue_features_train.csv")
dengue
```

```python
dengue.isna().sum()
```

```python
types = dict(dengue.dtypes)
```

En las columnas numéricas, cambiamos las entradas vacías con la media de la columna

```python
for column in types:
    if types[column] in [int, float]:
        dengue[column].fillna(dengue[column].mean(), inplace=True)
```

En este caso, el único valor categórico es city. De modo que vamos a pasarlo a numérico.

```python
for city in set(dengue['city']):
    dengue["city_" + city] = dengue["city"].apply(lambda x: x == city) * 1 # Multiplicamos por 1 para que aparezca 1 o 0
```

Y borramos la columna original

```python
dengue.drop("city", inplace=True, axis=1)
```

Y borramos la columna de la fecha

```python
dengue.drop("week_start_date", inplace=True, axis=1)
```

Quedando nuestro dataset de la siguiente manera:

```python
dengue
```

Finalmente, guardamos este dataset como csv para su posterior análisis

```python
dengue.to_csv("dengue_limpio.csv", index=False)
```
